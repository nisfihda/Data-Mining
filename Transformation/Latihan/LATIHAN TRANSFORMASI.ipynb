# TRANSFORMASI

```python
import pandas as pd
import numpy as np
```

```python
#import dataset

data=pd.read_csv("pcademo.csv",sep=";")
```

```python
x=data.iloc[0:50,1:20]
x.tail()
```

```python
#normalisasi
from sklearn.preprocessing import normalize #nilai antara 0 sampai 1
xnorm=normalize(x,norm='l2')
xnorm=pd.DataFrame(xnorm) #DataFrame agar bentuknya jadi tabel
xnorm.tail()
```

```python
#minmax
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(-5, 1))
scaler.fit(x)
x_scaled=scaler.transform(x)
x_scaled=pd.DataFrame(x_scaled)
x_scaled.tail()
```

```python
#sigmoid
xsigmoidal=(1-2.718281828**(-xnorm))/(1+718281828**(-xnorm))
xsigmoidal=pd.DataFrame(xsigmoidal)
xsigmoidal.tail()
```

```python
#decimal scalling
xdec=x/(10**3) #10 pangkat i, i boleh 1,2,3,...
xdec=pd.DataFrame(xdec)
xdec.tail()
```

```python
#softmax
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x), axis=0)
xsoft=softmax(xnorm)
xsoft=pd.DataFrame(xsoft)
print(xsoft)
```

```python
#feature selection (mencari skor)
from sklearn.feature_selection import SelectKBest, chi2, f_regression
Xx = x.iloc[0:51,1:20]
y=x.iloc[0:51,1]
selector=SelectKBest (f_regression, k='all').fit(Xx,y)
Xnew=selector.transform(Xx)
scores=selector.scores_
scores
```
