# Transformasi
## Data Australia Open 2013

```python
import pandas as pd
import numpy as np
data=pd.read_csv("Australia Open 2013.csv", sep=";")
np.sum(data.isnull())
```

```python
data['DBF']=data['DBF'].fillna((data['DBF'].mean()))
data['NPW']=data['NPW'].fillna((data['NPW'].mean()))
np.sum(data.isnull())
```

```python
data.to_csv('Australia Open 2013_1.csv', sep=';')
```

```python
from scipy.stats import stats
z = np.abs(stats.zscore(data._get_numeric_data()))
print(z)
data2 = data[(z<3).all(axis=1)]
print(data2.shape)
```

```python
data2.to_csv('Australia Open 2013_2.csv', sep=';')
```

```python
#normalization untuk data tanpa missing value
from sklearn.preprocessing import normalize
norm = normalize(data,norm = 'l2')
norm = pd.DataFrame(norm)
norm.tail()
```

```python
#transformasi untuk data tanpa missing value
tsig1=(1-2.718281828**(-norm))/(1+2.718281828**(-norm))
tsig1=pd.DataFrame(tsig1)
tsig1.tail()
```

```python
tsig1.to_csv('Australia Open 2013_3.csv', sep=';')
```

```python
#normalization untuk data tanpa missing value dan outlier
from sklearn.preprocessing import normalize
norm2= normalize(data2,norm= 'l2')
norm2= pd.DataFrame(norm2)
norm2.tail()
```
